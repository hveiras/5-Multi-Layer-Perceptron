```python
import numpy as np
```

```python
logits=np.random.choice(range(10),10)
```

```python
proba=np.exp(logits)
proba=proba/proba.sum()
proba
```

```python
logits=np.random.choice(range(1000),100)
from matplotlib import pyplot as plt
for i in np.arange(0,3,0.01):
    proba=np.exp(logits*i)
    proba=proba/proba.sum()
    proba
    plt.plot(proba)
plt.show()
```
